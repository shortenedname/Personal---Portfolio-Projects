```python
'''import libraries and loading in the original .csv file and looking at the head and filling in empty spaces if there are any.'''
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm

master_df = pd.read_csv('all_data.csv')
master_df.fillna(0)
master_df.head(32)
```

```python
''' create a master list of lists of all column data by row so that it can be cleaned and stripped into different lists based on what's what. I started to do it in a much worse way, by country, but then went thru notes and saw a mention of .ierrows(), used it,
then peaked a little and saw I was moving in the right direction instead of nesting several for loops. still might try another version w/ the nested loops for funsies.
 '''

new_list = []
for index, row in master_df.iterrows():
    new_list.append(list(row))



countries_name = []
year_of_data= []
avg_age_of_year = []
gdp_of_year = []

#code_holder is really just the year

# iterate through the list of lists to pull out country name, the year for the data of the country, the average life expectancy age, and the gdp of that year
for index, row in enumerate(new_list):
#     print(row)
    countries_name.append(row.pop(0))
    year_of_data.append(row.pop(0))
    avg_age_of_year.append(row.pop(0))
    gdp_of_year.append(row.pop(0))


#run print statements for tests and an if/else to see the lists are all equal in length
print("this is a list of country names:", countries_name)
print("this is a list of the year of the data of each country:", year_of_data)
print("this is a list of the avg life expectancy each year per country:", avg_age_of_year)
print("this is the gdp of the country that year:", gdp_of_year)
#another test
if len(countries_name) == len(year_of_data) == len(avg_age_of_year) == len(gdp_of_year):
    print ("YES")
else:
    print("NAUR")
#print(new_list)
#print("this is the og", new_list)   this is checking to see if the original list is properly stripped
```

```python
'''here we're going to create a new dictionary and set the keys as something close to if not the same as the column names from the .csv file. Then convert thwe dictionary into a pandas dataframe using the .DataFrame( ) pandas method. Single line comments
are just playing w/ dictionaries and trying to see/remember what's what w/ them'''

ultimate_dictionary = {"Country": countries_name, "Year": year_of_data, "Expectancy": avg_age_of_year, "GDP": gdp_of_year }
#print(ultimate_dictionary)

#for key in ultimate_dictionary:
    #print (ultimate_dictionary[key])
#ult_dict_items = ultimate_dictionary.items()
#print(ult_dict_items)

converted_df = pd.DataFrame(data=ultimate_dictionary)
converted_df["Year"] = converted_df["Year"].astype(int)
print(converted_df.head(15))
print(converted_df.info())
```

```python
'''Here I'm starting to make graphs. Im betting that as the gdp increases, life expectrancy increases so lets start with that'''

sns.set_theme(style="darkgrid")
sns.lineplot(x=year_of_data, y=avg_age_of_year, hue="Country", data=converted_df)
plt.title('Age Expectancy of Countries from 2000 - 2015')
```
